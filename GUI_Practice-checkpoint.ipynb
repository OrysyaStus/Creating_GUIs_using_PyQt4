```python
#Creating a Widget with a functional button by using a signal
import sys
from PyQt4 import QtGui

class PrettyWidget(QtGui.QWidget):
    
    def __init__(self):
        super(PrettyWidget, self).__init__()
        self.initUI()
        #can add more onwards, created User Interface
    
    def initUI(self):
        self.setGeometry(100,100,1000,500)
        #100,100 is associated with where you are moving the GUI
        #1000,500 is associated with the size of the GUI
        self.setWindowTitle('ProQinase Cancer Cell Line Explorer')
        
        self.btn = QtGui.QPushButton('Button', self)
        self.btn.resize(self.btn.sizeHint())
        #built in function which creates size of the button based on size of the widget
        self.btn.move(10,40)  
        #define position of the button
        self.btn.clicked.connect(self.Signal)
        #the button will be able to create a signal, where 'clicked' makes the signal
        
        self.lbl = QtGui.QLabel(self)
        self.lbl.move(10,10)
        self.lbl.setText('Slot')
        self.lbl.adjustSize()
        #the label will be the slot
        
        self.show()
    
    def Signal(self):
        self.lbl.setText('Signal Received')
        self.lbl.move(10,10)
        self.lbl.adjustSize()
    #must define a signal 'Signal' to which the button connects to
    #takes the label above and outputs signal received
def main():
    app = QtGui.QApplication(sys.argv)
    w = PrettyWidget()
    app.exec_()
    #for running the widget
    #sys.exit(exec())      to exit out of the system

if __name__ == '__main__':
    main()
#setting up the environment for the python code
```

```python
#Creating a Widget using signal dialog, creates a text box
import sys
from PyQt4 import QtGui

class PrettyWidget(QtGui.QWidget):
    
    def __init__(self):
        super(PrettyWidget, self).__init__()
        self.initUI()
        #can add more onwards, created User Interface
    
    def initUI(self):
        self.setGeometry(100,100,1000,500)
        #100,100 is associated with where you are moving the GUI
        #1000,500 is associated with the size of the GUI
        self.setWindowTitle('ProQinase Cancer Cell Line Explorer')
        
        self.btn = QtGui.QPushButton('Button', self)
        self.btn.resize(self.btn.sizeHint())
        #built in function which creates size of the button based on size of the widget
        self.btn.move(10,40)  
        #define position of the button
        self.btn.clicked.connect(self.signalDialog)
        #the button will be able to create a signal, where 'clicked' makes the signal
        
        self.lbl = QtGui.QLabel(self)
        self.lbl.move(10,10)
        self.lbl.setText('Slot')
        self.lbl.adjustSize()
        #the label will be the slot
        
        self.show()
    
    def signalDialog(self):
        text, ok = QtGui.QInputDialog.getText(self, 'Dialog Box', 'Enter text:' )
        if ok:
            self.lbl.setText(str(text))
            self.lbl.adjustSize()
            
def main():
    app = QtGui.QApplication(sys.argv)
    w = PrettyWidget()
    app.exec_()
    #for running the widget
    #sys.exit(exec())      to exit out of the system

if __name__ == '__main__':
    main()
#setting up the environment for the python code
```

```python
#all layouts above completed with absolute positioning, if the GUI becomes smaller the button position's does not adjust to resizing
#good to use instead box layout or grid layout

#1. horizontal box layout
#y coordinate adjusts and x coordinate stays the same
import sys
from PyQt4 import QtGui

class PrettyWidget(QtGui.QWidget):
    
    def __init__(self):
        super(PrettyWidget, self).__init__()
        self.initUI()
           
    def initUI(self):
        self.setGeometry(600,300,400,200)
        self.setWindowTitle('ProQinase Cancer Cell Line Explorer')
        
        hbox = QtGui.QHBoxLayout()
        hbox.addStretch(1)
        #the ratio of the change in the element coordinate:to siye of the window is a 1:1 stretch
        
        btn1 = QtGui.QPushButton('Button 1', self)
        btn1.resize(btn1.sizeHint())
        hbox.addWidget(btn1)
        
        btn2 = QtGui.QPushButton('Button 2', self)
        btn2.resize(btn2.sizeHint())
        hbox.addWidget(btn2)
        #stacks from the left
        
        self.setLayout(hbox)
        
        self.show()
    
def main():
    app = QtGui.QApplication(sys.argv)
    w = PrettyWidget()
    app.exec_()

if __name__ == '__main__':
    main()
```

```python
#combining horizontal and vertical box layouts
#be sure to draw it out first as it will make the coding a lot easier
import sys
from PyQt4 import QtGui

class PrettyWidget(QtGui.QWidget):
    
    def __init__(self):
        super(PrettyWidget, self).__init__()
        self.initUI()
           
    def initUI(self):
        self.setGeometry(600,300,400,200)
        self.setWindowTitle('ProQinase Cancer Cell Line Explorer')
        
        #VBOX1
        
        vbox1 = QtGui.QVBoxLayout()
        vbox1.addStretch(1)
        
        btn1 = QtGui.QPushButton('Button 1', self)
        btn1.resize(btn1.sizeHint())
        vbox1.addWidget(btn1)
        
        btn2 = QtGui.QPushButton('Button 2', self)
        btn2.resize(btn2.sizeHint())
        vbox1.addWidget(btn2)
        
        #VBOX2
        
        vbox2 = QtGui.QVBoxLayout()
        vbox2.addStretch(1)
        
        btn3 = QtGui.QPushButton('Button 3', self)
        btn3.resize(btn3.sizeHint())
        vbox2.addWidget(btn3)
        
        btn4 = QtGui.QPushButton('Button 4', self)
        btn4.resize(btn4.sizeHint())
        vbox2.addWidget(btn4)   
        #addWidget when adding a Widget object
        
        #HBOX
        
        hbox = QtGui.QHBoxLayout()
        hbox.addStretch(1)
        hbox.addLayout(vbox1)
        hbox.addLayout(vbox2)
        #addLayout when adding a Layout object
        
        self.setLayout(hbox)
        
        self.show()
    
def main():
    app = QtGui.QApplication(sys.argv)
    w = PrettyWidget()
    app.exec_()

if __name__ == '__main__':
    main()
```

```python
#run code above and then can run code below to allow for python plotting
import sys
from PyQt4 import QtGui

class PrettyWidget(QtGui.QWidget):
    
    def __init__(self):
        super(PrettyWidget, self).__init__()
        self.initUI()
        
        
    def initUI(self):
        self.setGeometry(600, 300, 1000, 600)
        self.center()
        self.setWindowTitle('Grid Plotting')     
        
        #using grid layout
       
        grid = QtGui.QGridLayout()
        self.setLayout(grid)
       
        btn1 = QtGui.QPushButton('Plot 1', self)
        btn1.resize(btn1.sizeHint()) 
        btn1.clicked.connect(self.plot1)
        grid.addWidget(btn1, 2,0)

        btn2 = QtGui.QPushButton('Plot 2', self)
        btn2.resize(btn2.sizeHint()) 
        btn2.clicked.connect(self.plot2)
        grid.addWidget(btn2, 2,1)
       
        self.figure = plt.figure(figsize=(15,5))
        #15,5 is the size
        self.canvas = FigureCanvas(self.figure)
        #put empty figure inside the canvas
        self.toolbar = NavigationToolbar(self.canvas, self)
        grid.addWidget(self.canvas, 1,0,1,2)
        #1,2 is the size
        grid.addWidget(self.toolbar, 0,0,1,2)
               
        self.show()

    def plot1(self):
        plt.cla()
        #clears out plot from the canvas
        ax = self.figure.add_subplot(111)
        x = [i for i in range(100)]
        y = [i**2 for i in x]
        ax.plot(x,y, 'b.-')
        ax.set_title('Quadratic Plot')
        self.canvas.draw()

    def plot2(self):
        plt.cla()
        ax = self.figure.add_subplot(111)
        x = [i for i in range(100)]
        y = [i**0.5 for i in x]
        ax.plot(x,y, 'r.-')
        ax.set_title('Square Root Plot')
        self.canvas.draw()
        
    def center(self):
        qr = self.frameGeometry()
        cp = QtGui.QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())  

def main():
    app = QtGui.QApplication(sys.argv)
    w = PrettyWidget()
    app.exec_()


if __name__ == '__main__':
    main()
```

```python
#run the code above first
class PrettyWidget(QtGui.QWidget):
    
    def __init__(self):
        super(PrettyWidget, self).__init__()
        self.initUI()
        
        
    def initUI(self):
        self.setGeometry(600, 300, 400,200)
        self.setWindowTitle('Table')     
        
        #using grid layout
        grid = QtGui.QGridLayout()
        self.setLayout(grid)
       
        #Data
        data = {'Kitty':['1','2','3','3'], 
                'Cat':['4','5','6','2'], 
                'Meow':['7','8','9','5'],
                'Purr':['4','3','4','8'],}
        #in the form of a dictionary
        
        #Create Empty 5x5 table
        table = QtGui.QTableWidget(self)
        table.setRowCount(5)
        table.setColumnCount(5)
        
        #Enter data onto Table
        horHeaders = []
        for n, key in enumerate(sorted(data.keys())):
            horHeaders.append(key)
            for m, item in enumerate(data[key]):
                newitem = QtGui.QTableWidgetItem(item)
                table.setItem(m, n, newitem)
        #adding in data from above into empty table
        #where m=row, n=column name, newitem=values
        
        #Add Header
        table.setHorizontalHeaderLabels(horHeaders)
        
        #Adjust size of Table
        table.resizeColumnsToContents()
        table.resizeRowsToContents()
        #adjusts size of table to make it look nice
        
        #Add Table to Grid
        grid.addWidget(table,0,0)
        
        self.show()
        #can manipulate the table in real time

def main():
    app = QtGui.QApplication(sys.argv)
    w = PrettyWidget()
    app.exec_()


if __name__ == '__main__':
    main()
```

```python
#revision on plots, tables, and file browser
from PyQt4 import QtGui
import os, sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_qt4agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.backends.backend_qt4agg import NavigationToolbar2QTAgg as NavigationToolbar
```

```python
#execute code above first
class PrettyWidget(QtGui.QWidget):
    
    def __init__(self):
        super(PrettyWidget, self).__init__()
        self.initUI()
        
        
    def initUI(self):
        self.setGeometry(600, 300, 1000, 600)
        self.center()
        self.setWindowTitle('Revision on Plots, Tables, and File Browser')     
        
        #Grid Layout
        grid = QtGui.QGridLayout()
        self.setLayout(grid)
        
        #Canvas and Toolbar
        self.figure = plt.figure(figsize=(15,5))
        self.canvas = FigureCanvas(self.figure)
        self.toolbar = NavigationToolbar(self.canvas, self)
        grid.addWidget(self.canvas, 2,0,1,2)
        grid.addWidget(self.toolbar, 1,0,1,2)
        
        #Empty 5x5 table
        self.table = QtGui.QTableWidget(self)
        self.table.setRowCount(1)
        self.table.setColumnCount(9)
        grid.addWidget(self.table, 3,0,1,2)
        
        #Import CSV Button
        btn1 = QtGui.QPushButton('Import CSV', self)
        btn1.resize(btn1.sizeHint())
        btn1.clicked.connect(self.getCSV)
        grid.addWidget(btn1, 0,0)
        
        #Plot Button
        btn2 = QtGui.QPushButton('Plot', self)
        btn2.resize(btn2.sizeHint())
        btn2.clicked.connect(self.plot)
        grid.addWidget(btn2, 0,1)
        
        self.show()
        
    def getCSV(self):
        filePath = QtGui.QFileDialog.getOpenFileName(self,
                                                        'Multiple File',
                                                        "C:\Users\stor0616\Desktop",
                                                        '*.csv')
        fileHandle = open(filePath, 'r')
        line = fileHandle.readline()[:-1].split(',')
        #all the value up to the 2nd to last value b/c don't want to include /n in code
        for n, val in enumerate(line):
            newitem = QtGui.QTableWidgetItem(val)
            self.table.setItem(0, n, newitem)
        self.table.resizeColumnsToContents()
        self.table.resizeRowsToContents()
        #if all the values were numbers
    
    def plot(self):
        y = []
        for n in range(9):
            #knew that there were 9 values in the csv file
            try:
                y.append(float(self.table.item(0, n).text()))
            except:
                y.append(np.nan)
            #if entered a number then becomes float, if not become nan from numpy library
        plt.cla()
        ax = self.figure.add_subplot(111)
        ax.plot(y, 'r.-')
        ax.set_title('Table Plot')
        self.canvas.draw()
    
    def center(self):
        qr = self.frameGeometry()
        cp = QtGui.QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())
    
def main():
    app = QtGui.QApplication(sys.argv)
    w = PrettyWidget()
    app.exec_()


if __name__ == '__main__':
    main()
```
